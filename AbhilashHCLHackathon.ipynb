```python
df_loyalty_rules = df_loyalty_rules.tail(4)[['points_per_unit_spend', 'minimum_spend_threshold', 'bonus_point']]

newInd = ['bronze', 'silver', 'gold', 'platinum']
df_loyalty_rules.index = newInd
df_loyalty_rules.columns = ['points_per_unit_spend', 'minimum_spend_threshold', 'bonus_point']
df_loyalty_rules['points_per_unit_spend'] /= 10
df_loyalty_rules.loc['silver', 'bonus_point'] = 50
df_loyalty_rules.loc['bronze', 'bonus_point'] = 10
df_loyalty_rules.loc['platinum', 'bonus_point'] = 150
df_loyalty_rules.to_csv('/content/loyalty_rules_processed.csv', index=False)
df_loyalty_rules
```

```python
df_loyalty_rules['points_per_unit_spend'].fillna(df_loyalty_rules['points_per_unit_spend'].mode(), inplace=True)
```

Part 3

```python
new_df = pd.merge(df_trans_table, df_cust_details, on="customer_id", how="inner").groupby(['transaction_date','store_id', 'customer_id',  'loyalty_status'])['total_amount'].sum().reset_index()
new_df
```

```python
def f(df1, df2):
  ls = df1['loyalty_status']
  ta = df1['total_amount']
  ans = 0
  if ta < df2.loc[ls, 'minimum_spend_threshold']:
    return 0
  else:
    ans += df2.loc[ls, 'points_per_unit_spend']*ta
  if ta > 1000:
    ans += df2.loc[ls, 'bonus_point']
  return ans
```

```python
new_df['nlscore'] = new_df.apply(lambda row: f(row, df_loyalty_rules), axis=1)
```

```python
for cust_id, newLP in zip(new_df['customer_id'], new_df['nlscore']):
  df_cust_details.loc[df_cust_details['customer_id'] == cust_id, 'total_loyalty_points'] += newLP
```
